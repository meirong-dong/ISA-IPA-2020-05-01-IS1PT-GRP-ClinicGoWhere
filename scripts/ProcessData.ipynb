```python
import pandas as pd
import numpy as np
import json
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
#path 
path_full = "data\clinic_details.csv"
path_phpc = "data\phpc_clinic.csv"
```

```python
#get all the clinic list 
clinic_full = pd.read_csv (path_full)
clinic_phpc = pd.read_csv (path_phpc)

print("The shape of full clinic list is: ")
print(clinic_full.shape)
print("The shape of PHPC clinic list is: ")
print(clinic_phpc.shape)
```

```python
#amend wrong records 
clinic_full.loc[clinic_full.Sno == 768, "Address"] = "BLK 140 Corporation Drive #01-03 Singapore 610140"
clinic_full.loc[clinic_full.Sno == 2569, "Address"] = "3 Mt Elizabeth Mt Elizabeth Medical Centre #10-10 Singapore 228510"
#clinic_full.loc[clinic_full.Sno == 3011, "Address"] = "PULAU BUKOM Singapore 903808" not in main island
clinic_full.loc[clinic_full.Sno == 3488, "Address"] = "264 THOMSON ROAD #13-09 Singapore 574391"
clinic_full.loc[clinic_full.Sno == 3636, "Address"] = "BLK 273C Punggol Place #01-880 Singapore 823273"
```

```python
#remvoe all not found rows
clinic_list = [clinic_full,clinic_phpc]
for item in clinic_list:
    item[item.Name != "No Clinic Found"]
```

```python
clinic_full.head()
```

```python
#remove unwanted characters in the list
chars = ["\n","\t","[^\S\r\n]{2,}"]
for char in chars:
    if char == "[^\S\r\n]{2,}" :
        clinic_full["Service"] = clinic_full["Service"].replace(char, ",", regex=True)
    else:
        clinic_full["Service"] = clinic_full["Service"].replace(char, " ", regex=True)

clinic_list = [clinic_full,clinic_phpc]
for item in clinic_list:
    item["Name"] = item["Name"].replace('\s+', ' ', regex=True)
    
clinic_full["Address"] = clinic_full["Address"].replace("\s+", " ", regex=True) #remove mutiple white spaces
clinic_full["Opening Hours"] = clinic_full["Opening Hours"].replace("\s+", " ", regex=True) #remove mutiple white spaces
clinic_full["Contact Number"] = clinic_full["Contact Number"].replace(".0", "", regex=True) #remove .0
```

```python
clinic_full.head()
```

```python
#check for services or programs
df = clinic_full
df["24 Hour"] = df["Opening Hours"].str.contains("24 Hour").fillna('False').astype(np.int8) 
df["Medical"] = df["Service"].str.contains("General Medical" or "Specialist Medical").fillna(False).astype(int)    
df["Dental"] = df["Service"].str.contains("General Dental" or "Specialist Dental").fillna(False).astype(int) 
df["CDMP"] = df["Programmes"].str.contains("Chronic Disease Management Programme").fillna(False).astype(int)  
df["CHAS"] = df["Programmes"].str.contains("Community Health Assistance Scheme").fillna(False).astype(int)  
df["ISP"] = df["Programmes"].str.contains("Integrated Screening Programme").fillna(False).astype(int) 
df["School"] = df["Address"].str.contains("School").fillna(False).astype(int)
df["PHPC"] = df.Name.isin(clinic_phpc.Name).astype(np.int8)
df = df[df.School == 0] #remove school clinics
df = df.drop(["Check","Programmes","Fees","School"], axis=1) #remove unwanted columns
print(df.shape)
```

```python
#get the postal code
df["Postal"] = df["Address"].str[-6:]
print(df.shape)
```
